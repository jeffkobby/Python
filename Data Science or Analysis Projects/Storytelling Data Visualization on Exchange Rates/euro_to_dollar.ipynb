# Data Visualization on Exchange Rates

## Introducing the Dataset

The dataset describes Euro daily exchange rates between 1999 and 2022

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

We begin by importing the dataset into a pandas DataFrame.

```python
exchange_rates = pd.read_csv('euro-daily-hist_1999_2020.csv')
```

The goal of this project is to focus and visualize the exchange rate between the Euro and the American Dollar.

First, we will rename the 'US dollar' and Period\Unit: columns to 'US_dollar' and 'Time' respectfully

Then we will change the Time column to a datetime datatype and sort in ascending order

```python
exchange_rates.rename(columns={'Period\\Unit:': 'Time', '[US dollar ]': 'USD'}, inplace=True)
exchange_rates.columns
```

```python
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.tail()
```

```python
exchange_rates.reset_index(drop=True, inplace=True)
exchange_rates.sort_values(by='Time', inplace=True)
exchange_rates.head()
```

Now we isolate the Time and the USD columns

```python
euro_to_dollar = exchange_rates[['Time', 'USD']].copy()
```

```python
print(euro_to_dollar.head())
print(euro_to_dollar.info())
```

Change the data type of the USD column to float

```python
euro_to_dollar['USD'].value_counts()
```

Remove '-' values and convert values to float

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['USD'] != '-']
euro_to_dollar['USD'] = euro_to_dollar['USD'].astype(float)
```

## Rolling Mean
Depending on our goals, we may not want to show that daily variation on our graph. If we want to hide it and show only the long-term trends, we can use the rolling mean (also known as the moving average).

We will calculate the rolling means for the US_dollar column using a moving window of 30 days. Add the rolling means to a new column named rolling_mean

```python
euro_to_dollar['rolling_mean'] = euro_to_dollar['USD'].rolling(30).mean()
euro_to_dollar.head(50)
```

```python
import matplotlib.style as style
%matplotlib inline

plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

## Euro-Dollar rate before and during the COVID pandemic

```python
before_covid = euro_to_dollar.copy()[(euro_to_dollar['Time'].dt.year >= 2016) & 
                                     (euro_to_dollar['Time'].dt.year <= 2019)]

during_covid = euro_to_dollar.copy()[euro_to_dollar['Time'].dt.year == 2020]

```

```python
before_covid.head()
```

```python
before_covid.tail()
```

```python
during_covid.head()
```

```python
# we will use the fivethirtyeight style for this plot
style.use('fivethirtyeight')

# add subplots
plt.figure(figsize=(10,5))
ax1 = plt.subplot()

ax1.plot(before_covid['Time'], before_covid['rolling_mean'], linewidth=1, color='green', label='Before Covid')
ax1.plot(during_covid['Time'], during_covid['rolling_mean'], linewidth=1, color='red', label='During Covid')
ax1.axvline(x=18260, ymin=0, ymax=0.9, linewidth=1, c='black', alpha=0.7)

# changes to all subplot
ax1.set_ylim(1, 1.25)
ax1.grid(alpha=0.5)
ax1.set_yticks([1.05, 1.10, 1.15, 1.20])
ax1.set_yticklabels(['1.05', '1.10','1.15', '1.20'], alpha=0.3)
ax1.set_xticklabels([])

years = ['2016', '2017', '2018', '2019', '2020', '2021']

# set x ticks
y=0.98
x=16745
for year in years:
    ax1.text(x, y, s=year, alpha=0.3)
    x += 365

# set title and subtitle
ax1.text(x=16600, y=1.28, s='The Euro-Dollar Rate Before And During The COVID-19 Pandemic', weight='bold', size=14)
ax1.text(x=16600, y=1.265, s='The USD depreciated quickly after a long period of steady appreciation due to the impact of COVID-19 on the US economy', size=10)

plt.show()
```
