## Exploring eBay Car Sales Data
For this project, I will be exploring a dataset of used cars from eBay Kleinanzeigen. The dataset was originally scraped and uploaded to Kaggle by user orgesleka. You can find the dataset  [here](https://data.world/data-society/used-cars-data).

### The data dictionary provided with the data is as follows:
dateCrawled - When this ad was first crawled. All field-values are taken from this date.

name - Name of the car.

seller - Whether the seller is private or a dealer.

offerType - The type of listing

price - The price on the ad to sell the car.

abtest - Whether the listing is included in an A/B test.

vehicleType - The vehicle Type.

yearOfRegistration - The year in which the car was first registered.

gearbox - The transmission type.

powerPS - The power of the car in PS.

model - The car model name.

kilometer - How many kilometers the car has driven.

monthOfRegistration - The month in which the car was first registered.

fuelType - What type of fuel the car uses.

brand - The brand of the car.

notRepairedDamage - If the car has a damage which is not yet repaired.

dateCreated - The date on which the eBay listing was created.

nrOfPictures - The number of pictures in the ad.

postalCode - The postal code for the location of the vehicle.

lastSeenOnline - When the crawler saw this ad last online

## Inital Exploration and Cleaning
I changed the "price", "odometer" and "yearOfRegistration" columns to int, int and str respectively

I also changed the column names of "price" to "price_$" and "odometer" to "odometer_km"

```python
autos.describe(include='all')
```

```python
# remove non-numeric characters and convert odomter and price columns to int
# convert registration_year column to string
autos['odometer'] = autos['odometer'].str.replace(',','').str.replace('km','').astype(int)
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(int)
```

## Exploring Price by Brand

```python
brands = autos['brand']
```

```python
brands.unique()
```

```python
brands.value_counts()[0:10]
```

```python
brands_count = brands.value_counts(normalize=True, dropna=False).sort_values(ascending=False)
print(brands_count[0:10])
```

```python
top_5_percent = brands_count[brands_count > .04]
print(top_5_percent)
```

```python
from pprint import pprint

average_price = {}

# autos[autos['brand'] == 'bmw']['price']

# loop through the top 5% brands list
for brand in top_5_percent.index:
    brand_df = autos[autos['brand'] == brand]
    average_price[brand] = brand_df['price'].mean()

pprint(average_price)
```

1. Audi, Mercedes Benz and BMW have the highest average price because they are luxury brands

2. Ford, Opel and Renault are make relatively cheap cars

3. Volkswagen cars are the most popular cars in out dataset. The cars they make are averagely neither cheap nor expensive

# Storing Aggregate Data in a DataFrame

## Using the groupby() method

I used this method to group data by brands and then calculate the average mileage for every car made by the brand on Ebay

```python
groups = autos.groupby('brand')
average_mileage = {}

# loop through the top 5% brands list
for brand in top_5_percent.index:
    brand_df = groups.get_group(brand)
    average_mileage[brand] = brand_df['odometer_km'].mean()

pprint(average_mileage)
```

```python
mean_mileage = pd.Series(average_mileage)
mean_price = pd.Series(average_price)
print(mean_mileage)
print(mean_price)
```

```python
mean_df = pd.DataFrame(mean_price, columns=['mean_price'])
mean_df
```

```python
mean_df['mean_mileage'] = mean_mileage
```

```python
mean_df
```

### Obersvations
The mean mileage for the top 5% of car brands is withing the 125000km-150000km range. However it is important to note that the data above is probably skewed due to the high number of cars with an odometer reading >150000km
