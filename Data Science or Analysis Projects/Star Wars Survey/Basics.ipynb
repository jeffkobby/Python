# Cleaning and Mapping Yes/No Columns
The answers in these columns can either be Yes or No; however, the can also be NaN values where the respondent chooses not to answer a question

```python
print(star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts())
print(f"Number of null values: {star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].isnull().sum()}")
```

```python
print(star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts())
print(f"Number of null values: {star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].isnull().sum()}")
```

```python
# create a dictionary to map values in the series to new values
map_values = {
    "Yes": True,
    "No": False
}

star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(map_values)
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(map_values)
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

# Cleaning and Mapping Checkbox Columns

The next six columns represent a single checkbox question. For each of these columns, if the value of the cell is the name of a movie, that means the respondent saw that movie. If the value is NaN, the respondent did not see the movie

```python
# rename the columns to names that are more intuitive
checkbox_cols = star_wars.columns[3:9]
count = 1
new_col_name = {}

for col in checkbox_cols:
    new_col_name[col] = f"seen_{count}"
    count += 1

print(new_col_name)
star_wars.rename(columns=new_col_name, inplace=True)
```

```python
# create a mapping dictionary
bool_map = {}

for col in star_wars.columns[3:9]:
#     get unique value for each series
    unique_values = star_wars[col].unique() 
    movie_name = list(unique_values)[0]
    bool_map[movie_name] = True

bool_map[np.NaN] = False

# apply mapping on columns
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(bool_map)

star_wars.head()
```

# Cleaning the Ranking Columns
The next six columns ask the respondent to rank the Star Wars movies in order from least to most favourite (1 means most favourite and 6 means least favourite).

Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film. - How much the respondent liked Star Wars: Episode I The Phantom Menace

Unnamed: 10 — How much the respondent liked Star Wars: Episode II Attack of the Clones

Unnamed: 11 — How much the respondent liked Star Wars: Episode III Revenge of the Sith

Unnamed: 12 — How much the respondent liked Star Wars: Episode IV A New Hope

Unnamed: 13 — How much the respondent liked Star Wars: Episode V The Empire Strikes Back

Unnamed: 14 — How much the respondent liked Star Wars: Episode VI Return of the Jedi

I will convert each column to a numeric type and rename the column

```python
# convert the columns to a float type
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# give the ranking columns a more descriptive name
ranking_cols = star_wars.columns[9:15]
count = 1
new_ranking_cols = {}

for col in ranking_cols:
    new_ranking_cols[col] = f"ranking_{count}"
    count += 1
    
# rename the columns in the dataframe
star_wars.rename(columns = new_ranking_cols, inplace=True)


star_wars.iloc[:, 9:15]
```

# Finding the Highest Ranked Movie
To find the highest ranking movie, I will use take the mean of each of the ranking columns in the dataframe

```python
star_wars.describe()
```

```python
# create a dataframe from the ranking columns
ranking = star_wars[star_wars.columns[9:15]].mean().sort_values()
```

```python
# Note: Ranking value range is 1-6 where 1 is the highes and
ranking.plot(kind='bar', title='Star Wars Average Movie Ranking', rot=20)
plt.xlabel('Movie')
plt.ylabel('Average Ranking Value')
plt.show()
```

## Summary
From the bar chart, we can see that the 5th movie in the series (Star Wars: Episode V The Empire Strikes Back) was the highest ranked movie. Bear in mind that the lower average value ranking, the higher ranked the movie is based on the survey.

# Finding the Most Viewed Movie
We will use the "seen" columns I renamed earlier in this project to find the most movies. Since True and False can be considered as 1 and 0 respectively, I will use the df.sum() method to determine the most viewed movie.

```python
# create a dataframe based on the seen columns
seen_cols = star_wars[star_wars.columns[3:9]]
seen_cols.head()
```

```python
most_viewed = seen_cols.sum().sort_values(ascending=False)
most_viewed
```

```python
most_viewed.plot.bar(title='Most Viewed Movie')
plt.xlabel('Movie')
plt.ylabel('View Count')
plt.show()
```

## Observation
When we compare the ranking and most viewed charts, we see that the ranking of a movie in this survey correlates with the number of respondents who have seen the movie. The highest ranked movie (ranking_5) is also the most viewed movie (seen_5) while the lowest ranked movie (ranking_3) is the least viewed movie (seen_3)

# Exploring Data in Binary Segments
In this section, I will group the respondents into two group (male and female) and perform the two previous analyses (highest ranked movie and most viewed movie) for each group

```python
males = star_wars[star_wars['Gender'] == 'Male']
females = star_wars[star_wars['Gender'] == 'Female']
```

```python
males.head()
```

```python
females.head()
```

# Highest ranked movie for each gender

```python
# find the mean male ranking for each movie
plt.figure(figsize=(10,3))

male_ranking = males[males.columns[9:15]].mean().sort_values()
female_ranking = females[females.columns[9:15]].mean().sort_values()

plt.subplot(1,2,1)
male_ranking.plot(kind='bar', title='Higest Ranked Movie - Male', rot=20)
plt.xlabel('Movie')
plt.ylabel('Rank Value')

plt.subplot(1,2,2)
female_ranking.plot(kind='bar', title='Higest Ranked Movie - Female', rot=20)
plt.xlabel('Movie')
plt.ylabel('Rank Value')

plt.show()
```

It is important to note that the lower the average ranking value, the higher a movie is ranked. From the chart, we can see that the fifth movie was the highest ranked movie according to the survey whilst the third movie was the lowest ranked for both genders.

# Most Viewed Movie for each gender

```python
male_seen_cols = males[males.columns[3:9]]
female_seen_cols = females[females.columns[3:9]]

male_seen_cols.head()
```

```python
male_most_viewed = male_seen_cols.sum().sort_values()
female_most_viewed = female_seen_cols.sum().sort_values()

plt.figure(figsize=(10,3))

plt.subplot(1,2,1)
male_most_viewed.plot(kind='bar', rot=20, title='Most Viewed Movie - Male')
plt.yticks(ticks=[0,100,200,300,400])
plt.xlabel('Movie')
plt.ylabel('Number of Views')

plt.subplot(1,2,2)
female_most_viewed.plot(kind='bar', rot=20, title='Most Viewed Movie - Female')
plt.yticks(ticks=[0,100,200,300,400])
plt.xlabel('Movie')
plt.ylabel('Number of Views')

plt.show()
```

# Observation
Both genders show similar trends when it comes to the number of views per movie.

The third movie is the least viewed movie and the fifth movie is most viewed movie.
